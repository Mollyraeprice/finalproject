```python
# Import our dependencies
import pandas as pd
```

```python
#load SCV file to dataFrame
house_df = pd.read_csv("austinHousingCombined.csv")
print(house_df.shape)
house_df.head()
```

```python
house_df.dtypes
```

```python
#filtering the data to only include single family 
single_family_df = house_df[house_df["hometype"] == "Single Family"]
print(single_family_df.shape)
single_family_df.head()
```

```python
#checking for null values 
single_family_df.isnull().sum()
```

```python
#filtering the data to austin city 
austin_df = single_family_df[single_family_df["city"] == "austin"]
print(austin_df.shape)
austin_df.head()
```

```python
austin_df["numofcommunityfeatures"].value_counts()
```

```python
austin_df["numofaccessibilityfeatures"].value_counts()
```

```python
austin_df["numofwaterfrontfeatures"].value_counts()
```

```python
#dropping columns that have no real value to our analysis
austin_df.drop(columns=["numofphotos","numofcommunityfeatures","numofaccessibilityfeatures","numofwaterfrontfeatures","latestpricesource"],inplace=True)
```

```python
#dropping the hometype since we already flitered the data to single family homes
austin_df = austin_df.drop(["hometype"], axis=1)
```

```python
#dropping the city since we already flitered the data to Austin
austin_df = austin_df.drop(["city"], axis=1)
```

```python
austin_df.dtypes
```

```python
#can we convert object to int?
austin_df.nunique()
```

```python
#removing columns that can't be convert to a numerical data 
austin_df = austin_df.drop(["homeimage","streetaddress","description"], axis=1)
```

```python
#change latest_saledate to date format
austin_df['latest_saledate'] = pd.to_datetime(austin_df['latest_saledate'])
```

```python
austin_df.dtypes
```

```python
#coverting bool values to int
austin_df.replace({True: 1, False:0}, inplace=True)
```
